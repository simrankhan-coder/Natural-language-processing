```python
import spacy
nlp=spacy.load('en_core_web_sm')
```

```python
s='Eshant has studied Data Science from books and is now making various projects'
d=nlp(s)
```

```python
for token in d:
  print(f"{token.text:{15}}{token.tag_:{15}}{spacy.explain(token.tag_)}")
```

##Visualization on Parts of Speech

```python
from spacy import displacy
displacy.render(d,jupyter=True)
```
