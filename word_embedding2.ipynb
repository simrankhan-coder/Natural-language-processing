##Importing the Libraries

```python
import gensim.downloader as api
from gensim.models import Word2Vec


import pandas as pd
import plotly.io as pio
import plotly.graph_objects as go
```

###Training the Model

```python
#print(api.info()['corpora'])
corpus=api.load('text8')
corpus=[word for word in corpus]

corpora=''
for cors in corpus:
  corpora+=' '.join(cors)
print('corpus size:',len(corpora.split(' ')))
print('Vocab size:', len(set(corpora.split(' '))))

model=Word2Vec(corpus,min_count=1000,vector_size=200,window=3,sg=0,epochs=5)
```

##Reducing the Dimensions

```python
from sklearn.decomposition import PCA

vec=[model.wv[word] for word in model.wv.key_to_index]
words=list(model.wv.key_to_index)

print('Len of word vector Before Dimension Reduction:',(vec[0]))

pca=PCA(n_components=2)
result=pca.fit_transform(vec)

print('Len of word vector After Dimension Reduction:',(result[0]))
```

```python
data={'X':result[:,0],'Y':result[:,1],'Label':words}
data=pd.DataFrame(data)
data.head()
```

###Visualising Word Vectors

```python
scatter=go.Scatter(x=data['X'],y=data['Y'],mode='markers+text',text=data['Label'],
                   textposition='top center',marker = dict( size = 5 , color = 'rgba(157,0,0, .8)'))
layout=go.Layout(title = 'Word Vectors Visualisation', xaxis = dict(title = 'Dimension - 1'), yaxis = dict(title = 'Dimension - 2'), hovermode = 'closest')
fig=go.Figure(data=scatter,layout=layout)
pio.show(fig)
```
