**Named Entity Recognition**

---


The named entity recognition (NER) is one of the most data preprocessing task. It involves the identification of key information in the text and classification into a set of predefined categories. An entity is basically the thing that is consistently talked about or refer to in the text.

```python
import spacy
nlp=spacy.load('en_core_web_sm')
```

```python
def show_ents(doc):
  if doc.ents:
    for ent in doc.ents:
      print(ent.text+' - '+ent.label_+' - '+str(spacy.explain(ent.label_)))
  else:
     print("No Entity Found")
```

```python
d=nlp(u"IIT is an Indian Institution which provides one of the finest education")
show_ents(d)
```

add a new entity

```python
d=nlp(u"Tesla is earning money at an extensive rate")
from spacy.tokens import Span as ss
ORG=d.vocab.strings[u"ORG"]
new_entity=ss(d,0,1,label=ORG)
# Get current entities as a mutable list
current_ents = list(d.ents)
# Filter out any entities that overlap with new_entity
filtered_ents = [ent for ent in current_ents if not (ent.start < new_entity.end and ent.end > new_entity.start)]
# Add the new entity
filtered_ents.append(new_entity)
# Assign the updated list back as a tuple
d.ents = tuple(filtered_ents)
show_ents(d)
```

add multiple words as an entity

```python
d=nlp(u"Playing cricket and football are both good for health")
from spacy.matcher import PhraseMatcher
m=PhraseMatcher(nlp.vocab)
phrase=['football','cricket']
patterns=[nlp(text) for text in phrase]
m.add('sports',None,*patterns)
found=m(d)
found
#This shows that we have found cricket and football and the indexes of those are also give
```
