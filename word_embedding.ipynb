```python
import gensim.downloader as api
from gensim.models import Word2Vec
```

```python
print(api.info()['corpora'])
```

```python
corpus=api.load('text8')
```

```python
corpus=[word for word in corpus]
corpora=''
for cors in corpus:
  corpora+=' '.join(cors)
```

```python
print('corpus size:',len(corpora.split(' ')))
print('Vocab size:', len(set(corpora.split(' '))))
```

```python
model=Word2Vec(corpus,min_count=6,vector_size=200,window=3,sg=0,epochs=5)
```

```python
model.wv.most_similar('india', topn = 10)
```

```python
model.wv.similarity('car','truck')
```

```python
model.wv.similarity('pen','truck')
```
